```python
def fn_sales_level(SalesTotal):
    retLevel = None
    if SalesTotal >= 11000:
        retLevel = "Level 2"
    elif SalesTotal >= 500:
         retLevel = "Level 1"
    else:
        retLevel = "Level 0"
    return retLevel
from pyspark.sql.functions import udf
```

## Other operatios


```python
from pyspark.sql.functions import concat, lit

dfPlusColumn = dfcustomer.withColumn("contact", concat(dfcustomer.email_address, lit(" "), dfcustomer.phone_number)).limit(5)

dfPlusColumn.display()
```

```python
from pyspark.sql.functions import count, col

dfgroupy = (dfcustomer
            .groupBy("country")
            .agg(count("customerID").alias("count"))
            .orderBy(col("count").desc()))

dfgroupy.display()
```

###JOIN  

```python
dfSales = spark.read.table("samples.bakehouse.sales_transactions")

dfSales.display()

dfJoin = dfcustomer.join(dfSales, on="customerID", how="inner").limit(10).orderBy(col("customerID").desc())

dfJoin.display()

dfJoin2 = (
    dfcustomer
    .join(dfSales, on="customerID", how="inner")
    .select("*")
    .limit(10)
)
dfJoin2.display()
```

##Cast Convert

```python
from pyspark.sql.functions import concat, lit

dfCast = dfSales.select(
    col("dateTime").cast("string"),
    col("customerID").cast("string"),
    col("franchiseID").cast("string"),
    concat(col("dateTime"), lit(" "), col("customerID"), lit(" "), col("franchiseID")).alias("concat")
).limit(3)
dfCast.display()
```

```python
dfconvertion = dfSales.select(
    dfSales["dateTime"].cast("string")
    , dfSales["customerID"].cast("string")
    , dfSales["franchiseID"].cast("string")
    , concat(dfSales["dateTime"], lit(" "), dfSales["customerID"], lit(" "), dfSales["franchiseID"]).alias("concat")
    ).limit(3)

dfconvertion.display()
```

```python
from pyspark.sql.types import StringType
from pyspark.sql.functions import col, lit

df_salesProduct = spark.read.table("samples.bakehouse.sales_transactions").groupBy("product").sum("totalPrice").withColumnRenamed("sum(totalPrice)", "totalSales").limit(10)

sales_level_udf = udf(fn_sales_level, StringType())

df_salesProduct = df_salesProduct.withColumn("salesLevel", sales_level_udf(col("totalSales")))
df_salesProduct = df_salesProduct.withColumn("test", lit("test")) ##adiciona coluna teste
df_salesProduct.display()
```
